```python
import albumentations as A
from albumentations.pytorch import ToTensorV2
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

# Function to display images
def display_image(image, title="Image"):
    plt.imshow(image)
    plt.title(title)
    plt.axis('off')
    plt.show()

```

```python
# Load an image
image_path = '1.png'
image = Image.open(image_path).convert('RGB')  # Ensure RGB format
image = np.array(image)

# Ensure the image has 3 channels
if image.ndim == 2:
    image = np.expand_dims(image, axis=-1)
if image.shape[-1] == 1:
    image = np.repeat(image, 3, axis=-1)


display_image(image)
```

```python
# Augmentation: Random Rotate upto 2 degrees
rotate_transform = A.Rotate(limit=2, p=1.0)
rotated_image = rotate_transform(image=image)['image']
display_image(rotated_image, title="Randomly Rotated Image")
```

```python
# Augmentation: Elastic Transform
elastic_transform = A.ElasticTransform(alpha=0.2, sigma=100.0, alpha_affine=0.3, p=1.0)
elastic_image = elastic_transform(image=image)['image']
display_image(elastic_image, title="Elastic Transformed Image")
```

```python
# Augmentation: Gaussian Noise
gauss_noise_transform = A.GaussNoise(var_limit=(10.0, 30.0), p=1.0)
gauss_noise_image = gauss_noise_transform(image=image)['image']
display_image(gauss_noise_image, title="Gaussian Noise Image")
```

```python
# Augmentation: Affine Transform
affine_transform = A.Affine(scale=(0.95, 1.05), translate_percent=(0.01, 0.01), shear=(-2, 2), p=1.0)
affine_image = affine_transform(image=image)['image']
display_image(affine_image, title="Affine Transformed Image")
```

```python
# Augmentation: Optical Distortion
optical_distort_transform = A.OpticalDistortion(distort_limit=0.03, shift_limit=0.03, p=1.0)
optical_distort_image = optical_distort_transform(image=image)['image']
display_image(optical_distort_image, title="Optical Distorted Image")
```

```python
# Augmentation: CLAHE
clahe_transform = A.CLAHE(clip_limit=2, tile_grid_size=(8, 8), p=1.0)
clahe_image = clahe_transform(image=image)['image']
display_image(clahe_image, title="CLAHE Image")
```

```python
# Augmentation: Perspective Transform
perspective_transform = A.IAAPerspective(scale=(0.01, 0.03), p=1.0)
perspective_image = perspective_transform(image=image)['image']
display_image(perspective_image, title="Perspective Transformed Image")
```

```python
# Augmentation: Random Brightness Contrast
rbright_transform = A.RandomBrightnessContrast(brightness_limit=0.1, contrast_limit=0.1, p=1.0)
rbright_image = rbright_transform(image=image)['image']
display_image(rbright_image, title="Random Brightness Contrast Image")
```

```python
# Augmentation: Grid Distortion
grid_transform = A.GridDistortion(num_steps=3, distort_limit=0.02, p=1.0)
grid_image = grid_transform(image=image)['image']
display_image(grid_image, title="Grid Distortion Image")
```

```python
# Augmentation: Hue Saturation Value
hsv_transform = A.HueSaturationValue(hue_shift_limit=10, sat_shift_limit=10, val_shift_limit=10, p=1.0)
hsv_image = hsv_transform(image=image)['image']
display_image(hsv_image, title="Hue Saturation Value Image")
```

```python
# Augmentation: Blur (Gaussian or Median)
blur_transform = A.OneOf([
    A.GaussianBlur(blur_limit=(1, 5), p=1.0),
    A.MedianBlur(blur_limit=3, p=1.0)
], p=1.0)
blur_image = blur_transform(image=image)['image']
display_image(blur_image, title="Blurred Image")
```

```python
# Augmentation: Normalize
normalize_transform = A.Normalize(mean=(0.5, 0.5, 0.5), std=(0.5, 0.5, 0.5))
normalized_image = normalize_transform(image=image)['image']

# Note: Normalize transforms the image to float32 and changes the value range.
# Rescale to [0, 1] for display purposes
normalized_image = (normalized_image - normalized_image.min()) / (normalized_image.max() - normalized_image.min())
display_image(normalized_image, title="Normalized Image")
```

```python
# increase dpis
from PIL import Image
import tempfile
import matplotlib.pyplot as plt

def set_image_dpi(file_path):
    im = Image.open(file_path)
    length_x, width_y = im.size
    factor = min(1, float(1024.0 / length_x))
    size = int(factor * length_x), int(factor * width_y)
    im_resized = im.resize(size, Image.ANTIALIAS)

    with tempfile.NamedTemporaryFile(delete=False, suffix='.png') as temp_file:
        temp_filename = temp_file.name
        im_resized.save(temp_filename, dpi=(300, 300))

    return temp_filename


# Example usage
new_file_path = set_image_dpi('1.png')
print(f"Resized image saved to: {new_file_path}")

plt.imshow(Image.open(new_file_path))
```
